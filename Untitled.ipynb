```python
import _pickle as cp
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X, y = cp.load(open('winequality-white.pickle', 'rb'))
```

```python
N, D = X.shape
N_train = int(0.8 * N)
N_test = N - N_train

X_train = X[:N_train]
y_train = y[:N_train]
X_test = X[N_train:]
y_test = y[N_train:]
```

# Handin 1

```python
x= [3,4,5,6,7,8,9]
yaxis = []
for i in range(3,10):
    yaxis.append(np.count_nonzero(y == i))
yaxis
sns.set_style("darkgrid")
plt.figure(figsize = (20,10))
fig = sns.barplot(x = x,y = yaxis, color = "blue", saturation = .3)
```

# Handin 2

```python
y_hat = np.mean(y_train)

E_train = np.mean((y_train - y_hat) ** 2)

E_test = np.mean((y_test - y_hat) ** 2)

print("Mean squared error for train data: ", E_train, "\n")
print("Mean squared error for test data: ", E_test)
```

# Handin 3

```python
mean_train = np.mean(X_train, axis = 0)
std_train = np.std(X_train, axis = 0)
mean_y = np.mean(y_train)
X_train = (X_train - mean_train) / std_train
X_test = (X_test - mean_train) / std_train
#y_train = (y_train - mean_y)
#y_test = (y_test - mean_y)

X_sqr = np.linalg.inv(np.matmul(X_train.transpose(), X_train))
w = np.matmul(np.matmul(X_sqr, X_train.transpose()), y_train)

y_hat_train = np.matmul(X_train, w)
y_hat_test = np.matmul(X_test, w)

E_train = np.mean((y_hat_train - y_train) ** 2)

E_test = np.mean((y_hat_test - y_test) ** 2)

print("Mean squared error for train data: ", E_train, "\n")
print("Mean squared error for test data: ", E_test)
```

# Handin 4

```python
n = 20
Etest = []
Etrain = []
naxis = []
while n <= 600:
    X_sqr = np.linalg.inv(np.matmul(X_train[:n].transpose(), X_train[:n]))
    w = np.matmul(np.matmul(X_sqr, X_train[:n].transpose()), y_train[:n])
    
    y_hat_train = np.matmul(X_train[:n], w)
    y_hat_test = np.matmul(X_test, w)
    
    E_train = np.mean((y_hat_train - y_train[:n]) ** 2)

    E_test = np.mean((y_hat_test - y_test) ** 2)
    
    Etrain.append(E_train)
    Etest.append(E_test)
    naxis.append(n)
    
    n += 20

plt.plot(naxis,Etest)
plt.plot(naxis,Etrain)
plt.show()
    
```
